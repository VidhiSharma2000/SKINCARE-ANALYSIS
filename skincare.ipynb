```python
import pandas as pd

df = pd.read_csv("cosmetics.csv")

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.scatterplot(x=df["Price"], y=df["Rank"])
plt.title("Price vs. Rank")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

sns.scatterplot(x=df["Price"], y=df["Rank"], alpha=0.7, color="blue")

plt.xlabel("Product Price ($)")
plt.ylabel("Customer Rank (Lower is Better)")
plt.title("Price vs. Customer Rank for Skincare Products")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

sns.scatterplot(x=df["Price"], y=df["Rank"], alpha=0.7, color="blue", s=80)

sns.regplot(x=df["Price"], y=df["Rank"], scatter=False, color="red")

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.xlabel("Product Price ($)")
plt.ylabel("Customer Rank (Lower is Better)")
plt.title("Price vs. Customer Rank for Skincare Products")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

top_brands = df["Brand"].value_counts().head(10)  # Get top 10 brands

plt.figure(figsize=(10,6))
sns.barplot(x=top_brands.values, y=top_brands.index, palette="Blues_r")

plt.xlabel("Number of Products")
plt.ylabel("Brand")
plt.title("Top 10 Best-Selling Skincare Brands")

plt.show()
```

```python
top_rated_brands = df.groupby("Brand")["Rank"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_rated_brands.values, y=top_rated_brands.index, palette="Greens_r")

plt.xlabel("Average Customer Rank")
plt.ylabel("Brand")
plt.title("Top 10 Highest Rated Skincare Brands")

plt.show()
```

```python
plt.figure(figsize=(10,6))

top_categories = df["Label"].value_counts().head(10)  # Get top 10 categories
sns.barplot(x=top_categories.values, y=top_categories.index, palette="Purples_r")

plt.xlabel("Number of Products")
plt.ylabel("Category")
plt.title("Top 10 Most Common Skincare Product Categories")

plt.show()
```

```python
avg_price_category = df.groupby("Label")["Price"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=avg_price_category.values, y=avg_price_category.index, palette="Reds_r")

plt.xlabel("Average Price ($)")
plt.ylabel("Category")
plt.title("Top 10 Most Expensive Skincare Categories")

plt.show()
```

```python
top_rated_categories = df.groupby("Label")["Rank"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_rated_categories.values, y=top_rated_categories.index, palette="Greens_r")

plt.xlabel("Average Customer Rank")
plt.ylabel("Category")
plt.title("Top 10 Highest Rated Skincare Categories")

plt.show()
```

```python
import matplotlib.pyplot as plt
import squarify

plt.figure(figsize=(12,6))

fig, ax = plt.subplots(1, 2, figsize=(16, 8))

top_selling = df["Brand"].value_counts(normalize=True).head(10) * 100
most_expensive = df.groupby("Brand")["Price"].mean().sort_values(ascending=False).head(10)
top_rated = df.groupby("Brand")["Rank"].mean().sort_values(ascending=False).head(10)

ax[0].set_title("Top 10 Best-Selling Skincare Brands (%)")
squarify.plot(sizes=top_selling.values, label=[f'{b}\n{v:.1f}%' for b, v in zip(top_selling.index, top_selling.values)], alpha=0.7, color=["#FF9999","#66B3FF","#99FF99","#FFCC99","#FFD700","#FF6347"], ax=ax[0])
ax[0].axis("off")

ax[1].set_title("Top 10 Most Expensive Skincare Brands")
squarify.plot(sizes=most_expensive.values, label=[f'{b}\n${v:.2f}' for b, v in zip(most_expensive.index, most_expensive.values)], alpha=0.7, color=["#B0C4DE","#FFA07A","#8A2BE2","#20B2AA","#FFD700","#DC143C"], ax=ax[1])
ax[1].axis("off")

plt.figure(figsize=(12,6))
plt.title("Top 10 Highest Rated Skincare Brands")
squarify.plot(sizes=top_rated.values, label=[f'{b}\n{v:.2f}★' for b, v in zip(top_rated.index, top_rated.values)], alpha=0.7, color=["#32CD32","#ADFF2F","#7FFF00","#9ACD32","#228B22","#6B8E23"])
plt.axis("off")

plt.show()
```
